```python
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

alltext = responses['What are things that make you glad to live in Grinnell?  '].dropna()
listtext = alltext.tolist()
# listtext_withoutstopwords = [i for i in listtext if i not in sw_nltk]
finaltextq1 = "".join(listtext)
# listtextoneword = finaltextq1.split(' ')

# wordcloud = WordCloud().generate(finaltextq1)
# plt.imshow(wordcloud, interpolation='bilinear')
# print(listtext_withoutstopwords)
# print(listtextoneword)
```

```python
wc = word_count(finaltextq1)
wcsorted = sorted(wc.items(), key=lambda x:x[1], reverse=True)
dict_wcsorted = dict(wcsorted)
print(dict_wcsorted)
```

```python
result = {key:value for (key, value) in dict_wcsorted.items() if value >= 15}

words = list(result.keys())
counts = list(result.values())

plt.bar(range(len(result)), counts, tick_label=words)
plt.rcParams['figure.figsize'] = [5, 10]
plt.show()
```

```python
import xlrd
import os
from string import punctuation, translate
from collections import Counter

filename = u'BABGsurveydata.xlsx'
sheet_no = 1  # To get the first sheet of the workbook
path = 'Data'
punctuation_map = dict((ord(c), u' ') for c in punctuation)

for filename in os.listdir(path):
   if filename.endswith('BABGsurveydata.xlsx'):
      print filename
      workbook = xlrd.open_workbook(filename)
      sheet = workbook.sheet_by_index(sheet_no)
      values = []
      for row in range(sheet.nrows):
          for col in range(sheet.ncols):
              c = sheet.cell(row, col)
              if c.ctype == xlrd.XL_CELL_TEXT:
                 cv = unicode(c.value)
                 wordlist = cv.translate(punctuation_map).split()
                 values.extend(wordlist)
                 numberWords = Counter(wordlist)
                 print sum(numberWords.values()), ' words for that column'

      count = Counter(values)
      print sum(count.values()), ' total words counted (from all columns)'
```
